```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
from matplotlib.path import Path
import matplotlib.patches as patches

fig, ax = plt.subplots()
ax.axis([0,10,-10,10])

x = np.arange(0,10,0.2)
y = x*np.sin(x)
s = np.cos(x) + 2
y1 = y - s
y2 = y + s

p, = ax.plot([],[],'r--')

xs = np.concatenate([x,x[::-1],[x[0]]])
ys = np.concatenate([y1,y2[::-1],[y2[0]]])
vs = list(zip(xs,ys))
cd=[1]
for i in range(len(vs)-1):
    if i < len(vs)-2:
        cd.append(2)
    else:
        cd.append(79)

path = Path(vs, cd)
patch = patches.PathPatch(path, facecolor='orange',edgecolor='none',alpha=0.1)

tx = ax.text(0,0,'')

def init():
    tx.set_text('start')
    tx.set_position((0, 0))
    ax.add_patch(patch)
    return tx,patch,

def update(i):
    p.set_data(x[:i], y[:i])
    xs = np.concatenate([x[:i],x[i::-1],[x[0]]])
    ys = np.concatenate([y1[:i],y2[i::-1],[y2[0]]])
    vs = list(zip(xs,ys))
    cd=[1]
    for i in range(len(vs)-1):
        if i < len(vs)-2:
            cd.append(2)
        else:
            cd.append(79)
    
    path = Path(vs, cd)
    patch = patches.PathPatch(path, facecolor='orange',edgecolor='none',alpha=0.2)
    ax.add_patch(patch)
    tx.set_position((2,2))
    
    return p,patch,
    
ani = animation.FuncAnimation(fig, update,init_func=init,frames=len(x),interval=100,blit=True)

plt.close()

#HTML(ani.to_html5_video())                                                # HTML5 Video 出力
HTML(ani.to_jshtml())                    
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
plt.style.use('ggplot')

fig = plt.figure(figsize=(10,8))
ax = plt.subplot()

x = np.arange(0,10,0.4)
y = x*np.sin(x)
s = np.cos(x) + 2
y1 = y - s
y2 = y + s

def update(i):
    ax.cla()
    ax.axis([0,10,-8,8])
    ax.plot(x,y,'r--',alpha=0.2)
    ax.plot(x[:i+1],y[:i+1],'r-',lw=3)
    ax.fill_between(x[:i+1],y1[:i+1],y2[:i+1],color='r',alpha=0.1)
    
    xs = np.concatenate([x[:i+1],x[i::-1]])
    ys = np.concatenate([y1[:i+1],y2[i::-1]])
#     ax.plot(x[:i+1],y1[:i+1],'r+')
#     ax.plot(x[:i+1],y2[:i+1],'r+')
    for j in range(len(x[:i+1])*2):    
        ax.text(xs[j],ys[j],str(j),fontsize=10,color='m')
        ax.plot(xs[j],ys[j],'r+')
    
    return fig,

ani = animation.FuncAnimation(fig, update,frames=len(x),interval=100,blit=True)
plt.close()
#HTML(ani.to_html5_video())                                                # HTML5 Video 出力
HTML(ani.to_jshtml())              
```
