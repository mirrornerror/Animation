### ArtistAnimation with plt.plot(x,y) + plt.plot(x1,y1)

```python
fig = plt.figure()

x = np.linspace(-5,5,100)
y = np.sin(x)

ims = []
for i in range(len(x)):
    ln = plt.plot(x,y, 'b--')
    gl = plt.plot(x[:i],y[:i],'m-')
    pt = plt.plot(x[i],y[i],'ro')
    ims.append(ln + gl + pt)

ani = animation.ArtistAnimation(fig,ims,interval=50)
#ani.save(filepath + "-test.gif", writer = "imagemagick") # save as gif
plt.close()
HTML(ani.to_html5_video())
```

### FuncAnimation on plt.figure()

```python
x = np.linspace(-5,5,100)
y = np.sin(x)

fig = plt.figure()
plt.axis([-5,5,-2,2])      # :[xmin,xmax,ymin,ymax]

b, = plt.plot([],[],'b--',animated=True) # 
m, = plt.plot([],[],'m-',animated=True)  # 
r, = plt.plot([],[],'ro',animated=True)  # 
f = plt.fill_between(x, y-1.2,y+1,alpha=0.2, facecolor='g')
plt.axis('off') # 


def update(i):
    b.set_data(x,y)           #  
    m.set_data(x[:i], y[:i])  # 
    r.set_data(x[i], y[i])    # )
    path = f.get_paths()[0]
    path.vertices[:i, 0] = x[:i]
    path.vertices[:i, 1] = y[:i]-1
    plt.fill_between(x, y-1.2,y+1,alpha=0.2, facecolor='g')
    
    return b,m,r

ani = animation.FuncAnimation(fig, update, interval=50)

plt.close()

#HTML(ani.to_html5_video())   # HTML5 Video
HTML(ani.to_jshtml())      # Javascript HTML
```
